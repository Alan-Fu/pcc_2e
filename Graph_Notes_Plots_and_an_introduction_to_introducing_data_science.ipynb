# Problem 2 Two in a row, three in a row, ....

Using the $\pi$ digits, make a scatter plot that plots a point if there are two of the same number in a row. Also drop a point if the same number comes up three in a row. 
The 'x-axis' should be the number being repeated, while the 'y-axis' should be the number times $10^n$ where $n$ is the number of repeats. 
i.e. $333$ plots 333 versus $3*10^3= 300$.

Start off by plotting only a few thousand numbers, it will get slow when plotting all million digits.

**BONUS** Store the indices of when a repeat and three-peat occur. In one plot, plot the index of the repitions versus the digit. I have a feeling that this could look really cool as a 'polar' plot.  

```python
from matplotlib.markers import MarkerStyle
a = np.linspace(1,101,101)
b = np.random.random(a.size)
resize = plt.gcf().set_size_inches

# plt.plot(a,b,'r-o')
plt.plot(a, b, color='red',linestyle='dashed',marker='o')
plt.xlabel('Numbers 1 through 100')
plt.ylabel('Random Numbers')
plt.title('Example Plot')
resize(8,6)
plt.show()

thetas = np.linspace(-3*np.pi,3*np.pi,101)
plt.polar(thetas,a, 'g-o')
resize(8,6)
plt.title('Spiral')
plt.show()
```
